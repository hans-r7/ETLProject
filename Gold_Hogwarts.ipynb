## ü•á Gold Layer

### üì¶ Import the Function

```python
import pandas as pd
GOLD_PATH = "./Medallion Architecture/Gold"
```

### üîÉ Load the Silver Layer Data

```python
df = pd.read_parquet('./Medallion Architecture/Silver/Silver.parquet')
```

### üëÄ Preview the Dataset 

```python
df.head()
```

### üìä Reports and Insights

```python
df.info()
```

### 1Ô∏è‚É£ üêæ Report of Students Owning More Than One Pet 

```python
pet_counts = df.groupby("Student Name")["Pet"].count().reset_index()
pet_counts = pet_counts.rename(columns={"Pet": "Pet Count"})

students_with_multiple_pets = pet_counts[pet_counts['Pet Count'] > 1]
students_with_multiple_pets
```

```python
students_with_multiple_pets.to_parquet(f"{GOLD_PATH}/Gold_Student_Pets_Over_One.parquet", index=False)
```

###  2Ô∏è‚É£ üèÜ Report of Total Points Earned by `House` 

```python
total_house_points = df.groupby("House")["Points Earned for House"].sum().reset_index()
total_house_points = total_house_points.rename(columns={"Points Earned for House": "Total House Points"})
total_house_points = total_house_points.sort_values(by="Total House Points", ascending=False)

total_house_points
```

```python
total_house_points.to_parquet(f"{GOLD_PATH}/Gold_House_Points.parquet", index=False)
```

### 3Ô∏è‚É£ ‚ú® Report of Average Magical Accidents by `Defense Against the Dark Arts Grade`

```python
df["Defense Against the Dark Arts Grade"].unique()
```

```python
avg_magical_accidents = df.groupby("Defense Against the Dark Arts Grade")["Magical Accidents"].mean().reset_index()
avg_magical_accidents = avg_magical_accidents.rename(columns={"Magical Accidents": "Average Magical Accidents"})

avg_magical_accidents
```

```python
avg_magical_accidents.to_parquet(f"{GOLD_PATH}/Gold_Dark_Arts_Accidents.parquet", index=False)
```

###  4Ô∏è‚É£ üßô‚Äç‚ôÇÔ∏è Report of Students Participating in the Triwizard Tournament by `House`

```python
triwizard_tournament_involvement = df[df['Triwizard Tournament Involvement'] == 'Yes']
triwizard_by_house = triwizard_tournament_involvement.groupby('House')['Triwizard Tournament Involvement'].count().reset_index()

triwizard_by_house
```

```python
triwizard_by_house.to_parquet(f"{GOLD_PATH}/Gold_Triwizard_Tournament_House.parquet", index=False)
```

### 5Ô∏è‚É£ üßπ Report of Quidditch Positions by `House`

```python
quidditch_positions_by_house = df.groupby(['House', 'Quidditch Position']).size().reset_index(name='Count of Students')
quidditch_positions_by_house
```

### üíæ Save Reports as Parquet Files

```python
quidditch_positions_by_house.to_parquet(f"{GOLD_PATH}/Gold_Quidditch_House_Positions.parquet", index=False)
```
